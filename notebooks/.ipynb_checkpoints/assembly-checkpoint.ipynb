```python
import geopandas as gpd
```

```python
df = gpd.read_file('../data/riskscore/homicides.csv')
df.head(10)
```

```python
type(df)
```

```python
shapes = gpd.read_file('../data/geometries/zipcode_geometries.csv')
```

```python
shapes = shapes.rename(columns={'the_geom':'SHAPES',
                       'ZIP': 'ZIP_CD'})
```

```python
working_shapes = shapes[['ZIP_CD', 'SHAPES']]
```

```python
working_shapes['ZIP_CD'] = working_shapes['ZIP_CD'].astype('object')
```

```python
working_df = df[['ZIP_CD', 'ID']]
```

```python
working_df['ZIP_CD'] = working_df['ZIP_CD'].astype('object')
```

```python
working_df
```

```python
# remember: .nunique() automatically gets counts
zipcode = working_df.groupby('ZIP_CD')['ID'].nunique().sort_values(ascending=False).to_frame()
```

```python
import pandas as pd
```

```python
zipcode = pd.merge(zipcode, working_shapes, on='ZIP_CD')
```

```python
zipcode
```

```python
zipcode = zipcode.drop(zipcode.index[0])
zipcode = zipcode.rename(columns={'ID':'Homicides'})
zipcode['Score'] = 0
zipcode.head()
```

```python
def get_percentile_score(dataframe, category):
    
    fifth = int(dataframe[category].quantile(.95))
    tenth = int(dataframe[category].quantile(.90))
    twentyfifth = int(dataframe[category].quantile(.75))
    fiftieth = int(dataframe[category].quantile(.50))

    score = list()
    for index in dataframe.index:
        if dataframe[category][index] > fiftieth:
            if dataframe[category][index] > twentyfifth:
                if dataframe[category][index] > tenth:
                    if dataframe[category][index] > fifth:
                        score.append(4)
                    else:
                        score.append(3)
                else:
                    score.append(2)
            else:
                score.append(1)
        else:
            score.append(0)
    return score
```

```python
homicide_scores = get_percentile_score(zipcode, 'Homicides')
# shooting_scores = get_percentile_score(zipcode, 'Shootings')
```

```python
def add_scores(dataframe, *argv):
    for arg in argv:
        for index in dataframe.index:
            dataframe['Score'][index] = dataframe['Score'][index] + arg[index-1]
    return dataframe
```

```python
zipcode = add_scores(zipcode, homicide_scores)
```

```python
zipcode.to_csv('../app/resources/scores/zipcode_riskscores.csv')
```
